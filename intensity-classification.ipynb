```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from scipy import ndimage
import os




path = '/home/cot12/Documents/ddsm/ddsm-samples/'
ren = os.listdir(path)
length = len(ren)

keypoints = np.zeros((length, 9))

for m in range(0, length):
    addition = os.path.join(path,ren[m])
    image = cv2.imread(addition,0)
    Beginning_image = image
    print (ren[m])
   

    def Image_check_x(image):
        p = image.shape[0]
        if ((p+3)/4) != int((p+3)/4):
            if ((p+2)/4) != int((p+2)/4):
                if ((p+1)/4) != int((p+1)/4):
                    return p
                else:
                    return p+1
            else:
                return p+2
        else:
            return p+3



    def Image_check_y(image):
        p = image.shape[1]
        if ((p+3)/4) != int((p+3)/4):
            if ((p+2)/4) != int((p+2)/4):
                if ((p+1)/4) != int((p+1)/4):
                    return p
                else:
                    return p+1
            else:
                return p+2
        else:
            return p+3



    
    
    q = int(Image_check_x(Beginning_image))
    u = int(Image_check_y(Beginning_image))
    beginning_image = cv2.resize(Beginning_image,(u,q))
    pyrlvl1 = beginning_image
    Intensity = np.sum(beginning_image[:,:])
    
    area1 = np.zeros((pyrlvl1.shape[0], pyrlvl1.shape[1]))
    for j in range(1, pyrlvl1.shape[0] -1):
        for k in range(1, pyrlvl1.shape[1]-1):
            if pyrlvl1[j,k]> 0:
                area1[j,k] = 1
    a = np.sum(area1)
    mean_Intensity = Intensity/a
    keypoints[m, :] = np.array([[Intensity, mean_Intensity]])
    
```
